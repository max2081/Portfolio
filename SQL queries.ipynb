## Example queries

### Query number of arrests made

```python
%sql select count(id) from crime where arrest='TRUE'
```

### Query number of crimes at 'Gas Station'

```python
%sql select distinct(primary_type), location_description from crime where location_description = 'GAS STATION'
```

### Query communities with name beginning with B

```python
%sql select community_area_name from census where community_area_name like 'B%'
```

### Query which schools in Community Areas 10 to 15 are healthy certified?

```python
%sql select name_of_school, community_area_number, healthy_school_certified \
from c_schools \
where community_area_number >= 10 and community_area_number <= 15 \
and healthy_school_certified = 'Yes'
```

### Query average school safety score

```python
%sql select avg(safety_score) from c_schools
```

### Query retrieve list: the top 5 Community Areas by average College Enrollment

```python
%sql select community_area_name, \
cast(round(avg(college_enrollment)) as int) as average_college_enrollment \
from c_schools group by community_area_name \
order by average_college_enrollment desc limit 5
```

### Sub-query: Community Area with lowest school Safety Score?

```python
%sql select community_area_name, safety_score \
from c_schools \
where safety_score = (select min(safety_score) from c_schools)
```

### Query without using JOIN operator: Find the Per Capita Income of the Community Area which has a school Safety Score of 1

```python
%sql select c_schools.community_area_name, census.per_capita_income, c_schools.safety_score \
from c_schools, census \
where c_schools.community_area_number = census.community_area_number \
and safety_score = 1
```
